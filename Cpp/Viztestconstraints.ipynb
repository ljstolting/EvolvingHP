Visualize the data for the derived dynamical constraints. We expect three things.

1) Oscillatory circuits that are unsatisfiale for either condition are never pyloric
2) Oscillatory circuits that are guaranteed for either condition will satisfy the corresponding ordering critereon(ia)
3) Oscillatory circuits that are not guaranteed for either condition, but not unsatisfiable for either are sometimes pyloric and sometimes not

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
unsatisfiable = np.loadtxt('unsatisfiable.dat')
guaranteed = np.loadtxt('guaranteed.dat')
pyloric = np.loadtxt('pyloric.dat')
allcriteria = np.loadtxt('allcircuits.dat')
```

```python
print(len(pyloric))
print(sum(pyloric))
```

```python
osc_criteria = allcriteria[np.where(allcriteria>=3)]
len(osc_criteria)
```

```python
hyp1 = np.where(unsatisfiable.any() != 0)[0]
print("Total unsatisfiable: ", len(hyp1))
pyloric_count = sum(pyloric[hyp1])
plt.bar([0,1],[len(hyp1)-pyloric_count,pyloric_count]) 
plt.show()
#expect to miss either 1 or two pyloric characteristics
print(osc_criteria[hyp1])
```

```python
# IT IS IMPOSSIBLE TO GET A CIRCUIT THAT IS GUARANTEED TO BE PYLORIC IN ANY LIMIT CYCLE
hyp2 = np.where(guaranteed.any() == 1)[0]
print("Total where some guaranteed: ", len(hyp2))
pyloric_count = sum(pyloric[hyp2])
plt.bar([0,1],[len(hyp2)-pyloric_count,pyloric_count]) 
plt.show()
#expect to satisfy either 1 or 2 (or 3) ordering criteria
print(osc_criteria[hyp2])
```

```python
hyp1or2 = np.concatenate((hyp1,hyp2))
hyp3 = np.arange(len(pyloric))
hyp3 = np.delete(hyp3,hyp1or2)
print("Total neither: ", len(hyp3))
pyloric_count = sum(pyloric[hyp3])
plt.bar([False,True],[len(hyp3)-pyloric_count,pyloric_count]) 
plt.show()
#expect some to be pyloric some not
plt.bar([3,4,5,6],[len(np.where(osc_criteria[hyp3]==3)[0]),len(np.where(osc_criteria[hyp3]==4)[0]),len(np.where(osc_criteria[hyp3]==5)[0]),len(np.where(osc_criteria[hyp3]==6)[0])])
plt.show()
```
